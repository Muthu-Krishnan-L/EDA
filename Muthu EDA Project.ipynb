## Converting the Columns for further analysis
   

Column Name = Value, Details = Amount with Euro symbol as prefix and suffix ‘K’ or ‘M’ indicating thousands and millions respectively. Required Output = Convert to Float after getting rid of currency symbol and suffix.

```python
def val(sal):
    if sal[-1] == 'K':
        values = pd.to_numeric(sal[1:-1])
        values = values * 10000
    else:
        values = pd.to_numeric(sal[1:-1])
        values = values * 1000000
    return values    
```

```python
#Converting K and M values using function val
df['Value'] = df['Value'].apply(val)
```

```python
df['Value']
```

```python
df['Value'].dtype
```

Column Name = 'Wage' Details = Amount with Euro symbol as prefix and suffix ‘K’ or ‘M’ indicating thousands and millions respectively. Required Output = Convert to Float after getting rid of currency symbol and suffix.


```python
df['Wage'] = df['Wage'].str.replace('€','').str.replace('M','').str.replace('K','').astype(float)
```

```python
df['Wage'] = df['Wage']*1000
```

```python
df
```

```python
df['Wage'].dtype
```

Column Name = 'Joined' Details = Year as a string, in some cases complete date as stringrespectively. Required Output = Convert to int with only year

```python
#Checking for Null values
df['Joined'].isnull().sum()/len(df)*100
```

```python
#filling Nan values with mode imputation
df['Joined'].fillna(df.Joined.mode()[0],inplace=True)
```

```python
#Checking for Null values after mode imputation
df['Joined'].isnull().sum()/len(df)*100
```

```python
df['Joined'] = pd.DatetimeIndex(df['Joined']).year.astype(int)
```

```python
#checking the type for Joined column
df['Joined'].dtype
```

```python
#4 Column Name = 'Contract Valid Until' Details = Date as a string Required Output = Convert to datetime type
```

Column Name = 'Height' Details = In inches with a quotation mark Required Output = Convert to Float with decimal points

```python
df['Height'] = df['Height'].str.replace("\'",'.').astype(float)
```

Column Name = 'Weight' Details = Contains the suffix lbs Required Output = Remove the suffix and convert to float

```python
df['Weight'] = df['Weight'].str.replace('lbs','').astype(float)
```

```python
df['Height'].dtypes
```

Column Name = 'Release Clause' Details = Amount with Euro symbol as prefix and suffix ‘K’ or ‘M’ indicating thousands and millions respectively. Required Output = Convert to Float after getting rid of currency symbol and suffix.

```python
df['Release Clause'].astype(object)
```

```python
c=[]
for i in df['Release Clause']:
    if isinstance(i,float)==False:
        if i[-1]=='M':
            m = float(i[1:-1])
            f = m*1000000
            c.append(f)
        elif i[-1]=='K':
            k = float(i[1:-1])
            j = k*1000
            c.append(j)
        else:
            j=float(i[1:])
            c.append(j)
    else:
        c.append(i)
m=np.array(c)
df['Release Clause']=m
df['Release Clause'].fillna(df['Release Clause'].mean(),inplace=True)
```
